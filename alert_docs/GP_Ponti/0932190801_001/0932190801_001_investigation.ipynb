```python
%load_ext wurlitzer
import numpy as np
import os
import stingray as st
import matplotlib.pyplot as plt
from astropy.io import fits
from xspec import *
plt.rcParams['figure.figsize'] = [15, 8]

lc1 = fits.open('0844101101_PN129_dt1.lc')
bkg1 = fits.open('0844101101_PN129_dt1_bkg.lc')
full_gti1 = []
for j in lc1[2].data:
    full_gti1.append([j[0],j[1]])
soft1 = fits.open('0844101101_PN129_dt1_0.2-2keV.lc')
sbkg1 = fits.open('0844101101_PN129_dt1_0.2-2keV_bkg.lc')
soft_gti1 = []
for j in soft1[2].data:
    soft_gti1.append([j[0],j[1]])
hard1 = fits.open('0844101101_PN129_dt1_2-12keV.lc')
hbkg1 = fits.open('0844101101_PN129_dt1_2-12keV_bkg.lc')
hard_gti1 = []
for j in hard1[2].data:
    hard_gti1.append([j[0],j[1]])
    
lc2 = fits.open('0932190801_PN1_dt1.lc')
bkg2 = fits.open('0932190801_PN1_dt1_bkg.lc')
full_gti2 = []
for j in lc2[2].data:
    full_gti2.append([j[0],j[1]])
soft2 = fits.open('0932190801_PN1_dt1_0.2-2keV.lc')
sbkg2 = fits.open('0932190801_PN1_dt1_0.2-2keV_bkg.lc')
soft_gti2 = []
for j in soft2[2].data:
    soft_gti2.append([j[0],j[1]])
hard2 = fits.open('0932190801_PN1_dt1_2-12keV.lc')
hbkg2 = fits.open('0932190801_PN1_dt1_2-12keV_bkg.lc')
hard_gti2 = []
for j in hard2[2].data:
    hard_gti2.append([j[0],j[1]])
    
```

```python
full_lc1 = st.Lightcurve(lc1[1].data.field('TIME'),lc1[1].data.field('RATE'),err=lc1[1].data.field('ERROR'),gti=full_gti1,input_counts=False)
full_bkg1 = st.Lightcurve(bkg1[1].data.field('TIME'),bkg1[1].data.field('RATE'),err=bkg1[1].data.field('ERROR'),gti=full_gti1,input_counts=False)

soft_lc1 = st.Lightcurve(soft1[1].data.field('TIME'),soft1[1].data.field('RATE'),err=soft1[1].data.field('ERROR'),gti=soft_gti1,input_counts=False)
soft_bkg1 = st.Lightcurve(sbkg1[1].data.field('TIME'),sbkg1[1].data.field('RATE'),err=sbkg1[1].data.field('ERROR'),gti=soft_gti1,input_counts=False)

hard_lc1 = st.Lightcurve(hard1[1].data.field('TIME'),hard1[1].data.field('RATE'),err=hard1[1].data.field('ERROR'),gti=hard_gti1,input_counts=False)
hard_bkg1 = st.Lightcurve(hbkg1[1].data.field('TIME'),hbkg1[1].data.field('RATE'),err=hbkg1[1].data.field('ERROR'),gti=hard_gti1,input_counts=False)

full_lc1.apply_gtis()
full_bkg1.apply_gtis()
soft_lc1.apply_gtis()
soft_bkg1.apply_gtis()
hard_lc1.apply_gtis()
hard_bkg1.apply_gtis()

full_lc2 = st.Lightcurve(lc2[1].data.field('TIME'),lc2[1].data.field('RATE'),err=lc2[1].data.field('ERROR'),gti=full_gti2,input_counts=False)
full_bkg2 = st.Lightcurve(bkg2[1].data.field('TIME'),bkg2[1].data.field('RATE'),err=bkg2[1].data.field('ERROR'),gti=full_gti2,input_counts=False)

soft_lc2 = st.Lightcurve(soft2[1].data.field('TIME'),soft2[1].data.field('RATE'),err=soft2[1].data.field('ERROR'),gti=soft_gti2,input_counts=False)
soft_bkg2 = st.Lightcurve(sbkg2[1].data.field('TIME'),sbkg2[1].data.field('RATE'),err=sbkg2[1].data.field('ERROR'),gti=soft_gti2,input_counts=False)

hard_lc2 = st.Lightcurve(hard2[1].data.field('TIME'),hard2[1].data.field('RATE'),err=hard2[1].data.field('ERROR'),gti=hard_gti2,input_counts=False)
hard_bkg2 = st.Lightcurve(hbkg2[1].data.field('TIME'),hbkg2[1].data.field('RATE'),err=hbkg2[1].data.field('ERROR'),gti=hard_gti2,input_counts=False)

full_lc2.apply_gtis()
full_bkg2.apply_gtis()
soft_lc2.apply_gtis()
soft_bkg2.apply_gtis()
hard_lc2.apply_gtis()
hard_bkg2.apply_gtis()

```

```python
from matplotlib.ticker import StrMethodFormatter, NullFormatter
emin = 2.5
emax = 10
fig, (ax1,ax2) = plt.subplots(2, 1, sharex=True, gridspec_kw={'height_ratios': [3, 1], 'hspace':0},figsize=(12,8))
ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_xlim(emin, emax)
ax1.set_ylabel('counts s$^{-1}$ keV$^{-1}$')
ax1.errorbar(Plot.x(), Plot.y(),Plot.yErr(), Plot.xErr(),ls=' ', color='k')
ax1.plot(Plot.x(),Plot.model(), color='r')
ax2.set_xscale('log')
ax2.set_xlim(emin, emax)
ax2.set_xlabel('Energy (keV)')
ax2.set_ylabel('Norm. Residual')
ax2.xaxis.set_major_formatter(StrMethodFormatter('{x:.0f}'))
ax2.xaxis.set_minor_formatter(StrMethodFormatter('{x:.0f}'))
ax2.errorbar(Plot.x(), (np.asarray(Plot.y())-np.asarray(Plot.model()))/Plot.yErr(),1,Plot.xErr(), ls=' ', color='k')
ax2.hlines(0,emin, emax,color='k', lw=0.5)
plt.savefig('0932190801_001_model_fit_plot.png')
```
