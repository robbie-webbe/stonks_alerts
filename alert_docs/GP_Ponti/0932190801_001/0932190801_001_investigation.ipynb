```python
%load_ext wurlitzer
import numpy as np
import os
import stingray as st
import matplotlib.pyplot as plt
from astropy.io import fits
from xspec import *
plt.rcParams['figure.figsize'] = [15, 8]

lc = fits.open('0932190801_PN1_dt1.lc')
bkg = fits.open('0932190801_PN1_dt1_bkg.lc')
full_gti = []
for j in lc[2].data:
    full_gti.append([j[0],j[1]])
soft = fits.open('0932190801_PN1_dt1_0.2-2keV.lc')
sbkg = fits.open('0932190801_PN1_dt1_0.2-2keV_bkg.lc')
soft_gti = []
for j in soft[2].data:
    soft_gti.append([j[0],j[1]])
hard = fits.open('0932190801_PN1_dt1_2-12keV.lc')
hbkg = fits.open('0932190801_PN1_dt1_2-12keV_bkg.lc')
hard_gti = []
for j in hard[2].data:
    hard_gti.append([j[0],j[1]])
    
lc1 = fits.open('0844101101_PN129_dt1.lc')
bkg1 = fits.open('0844101101_PN129_dt1_bkg.lc')
full_gti1 = []
for j in lc1[2].data:
    full_gti1.append([j[0],j[1]])
soft1 = fits.open('0844101101_PN129_dt1_0.2-2keV.lc')
sbkg1 = fits.open('0844101101_PN129_dt1_0.2-2keV_bkg.lc')
soft_gti1 = []
for j in soft1[2].data:
    soft_gti1.append([j[0],j[1]])
hard1 = fits.open('0844101101_PN129_dt1_2-12keV.lc')
hbkg1 = fits.open('0844101101_PN129_dt1_2-12keV_bkg.lc')
hard_gti1 = []
for j in hard1[2].data:
    hard_gti1.append([j[0],j[1]])
```

```python
full_lc = st.Lightcurve(lc[1].data.field('TIME'),lc[1].data.field('RATE'),err=lc[1].data.field('ERROR'),gti=full_gti,input_counts=False)
full_bkg = st.Lightcurve(bkg[1].data.field('TIME'),bkg[1].data.field('RATE'),err=bkg[1].data.field('ERROR'),gti=full_gti,input_counts=False)

soft_lc = st.Lightcurve(soft[1].data.field('TIME'),soft[1].data.field('RATE'),err=soft[1].data.field('ERROR'),gti=soft_gti,input_counts=False)
soft_bkg = st.Lightcurve(sbkg[1].data.field('TIME'),sbkg[1].data.field('RATE'),err=sbkg[1].data.field('ERROR'),gti=soft_gti,input_counts=False)

hard_lc = st.Lightcurve(hard[1].data.field('TIME'),hard[1].data.field('RATE'),err=hard[1].data.field('ERROR'),gti=hard_gti,input_counts=False)
hard_bkg = st.Lightcurve(hbkg[1].data.field('TIME'),hbkg[1].data.field('RATE'),err=hbkg[1].data.field('ERROR'),gti=hard_gti,input_counts=False)

full_lc.apply_gtis()
full_bkg.apply_gtis()
soft_lc.apply_gtis()
soft_bkg.apply_gtis()
hard_lc.apply_gtis()
hard_bkg.apply_gtis()

full_lc1 = st.Lightcurve(lc1[1].data.field('TIME'),lc1[1].data.field('RATE'),err=lc1[1].data.field('ERROR'),gti=full_gti1,input_counts=False)
full_bkg1 = st.Lightcurve(bkg1[1].data.field('TIME'),bkg1[1].data.field('RATE'),err=bkg1[1].data.field('ERROR'),gti=full_gti1,input_counts=False)

soft_lc1 = st.Lightcurve(soft1[1].data.field('TIME'),soft1[1].data.field('RATE'),err=soft1[1].data.field('ERROR'),gti=soft_gti1,input_counts=False)
soft_bkg1 = st.Lightcurve(sbkg1[1].data.field('TIME'),sbkg1[1].data.field('RATE'),err=sbkg1[1].data.field('ERROR'),gti=soft_gti1,input_counts=False)

hard_lc1 = st.Lightcurve(hard1[1].data.field('TIME'),hard1[1].data.field('RATE'),err=hard1[1].data.field('ERROR'),gti=hard_gti1,input_counts=False)
hard_bkg1 = st.Lightcurve(hbkg1[1].data.field('TIME'),hbkg1[1].data.field('RATE'),err=hbkg1[1].data.field('ERROR'),gti=hard_gti1,input_counts=False)

full_lc1.apply_gtis()
full_bkg1.apply_gtis()
soft_lc1.apply_gtis()
soft_bkg1.apply_gtis()
hard_lc1.apply_gtis()
hard_bkg1.apply_gtis()
```
