```python
%load_ext wurlitzer
import numpy as np
import os
import stingray as st
import matplotlib.pyplot as plt
from astropy.io import fits
from xspec import *
plt.rcParams['figure.figsize'] = [15, 8]

lc = fits.open('0916811401_PN3_dt1.lc')
bkg = fits.open('0916811401_PN3_dt1_bkg.lc')
full_gti = []
for j in lc[2].data:
    full_gti.append([j[0],j[1]])
soft = fits.open('0916811401_PN3_dt1_0.2-2keV.lc')
sbkg = fits.open('0916811401_PN3_dt1_0.2-2keV_bkg.lc')
soft_gti = []
for j in soft[2].data:
    soft_gti.append([j[0],j[1]])
hard = fits.open('0916811401_PN3_dt1_2-12keV.lc')
hbkg = fits.open('0916811401_PN3_dt1_2-12keV_bkg.lc')
hard_gti = []
for j in hard[2].data:
    hard_gti.append([j[0],j[1]])
    
lc2 = fits.open('0916811501_PN8_dt1.lc')
bkg2 = fits.open('0916811501_PN8_dt1_bkg.lc')
full_gti2 = []
for j in lc2[2].data:
    full_gti2.append([j[0],j[1]])
soft2 = fits.open('0916811501_PN8_dt1_0.2-2keV.lc')
sbkg2 = fits.open('0916811501_PN8_dt1_0.2-2keV_bkg.lc')
soft_gti2 = []
for j in soft2[2].data:
    soft_gti2.append([j[0],j[1]])
hard2 = fits.open('0916811501_PN8_dt1_2-12keV.lc')
hbkg2 = fits.open('0916811501_PN8_dt1_2-12keV_bkg.lc')
hard_gti2 = []
for j in hard2[2].data:
    hard_gti2.append([j[0],j[1]])
```

```python
full_lc = st.Lightcurve(lc[1].data.field('TIME'),lc[1].data.field('RATE'),err=lc[1].data.field('ERROR'),gti=full_gti,input_counts=False)
full_bkg = st.Lightcurve(bkg[1].data.field('TIME'),bkg[1].data.field('RATE'),err=bkg[1].data.field('ERROR'),gti=full_gti,input_counts=False)

soft_lc = st.Lightcurve(soft[1].data.field('TIME'),soft[1].data.field('RATE'),err=soft[1].data.field('ERROR'),gti=soft_gti,input_counts=False)
soft_bkg = st.Lightcurve(sbkg[1].data.field('TIME'),sbkg[1].data.field('RATE'),err=sbkg[1].data.field('ERROR'),gti=soft_gti,input_counts=False)

hard_lc = st.Lightcurve(hard[1].data.field('TIME'),hard[1].data.field('RATE'),err=hard[1].data.field('ERROR'),gti=hard_gti,input_counts=False)
hard_bkg = st.Lightcurve(hbkg[1].data.field('TIME'),hbkg[1].data.field('RATE'),err=hbkg[1].data.field('ERROR'),gti=hard_gti,input_counts=False)

full_lc.apply_gtis()
full_bkg.apply_gtis()
soft_lc.apply_gtis()
soft_bkg.apply_gtis()
hard_lc.apply_gtis()
hard_bkg.apply_gtis()

full_lc2 = st.Lightcurve(lc2[1].data.field('TIME'),lc2[1].data.field('RATE'),err=lc2[1].data.field('ERROR'),gti=full_gti2,input_counts=False)
full_bkg2 = st.Lightcurve(bkg2[1].data.field('TIME'),bkg2[1].data.field('RATE'),err=bkg2[1].data.field('ERROR'),gti=full_gti2,input_counts=False)

soft_lc2 = st.Lightcurve(soft2[1].data.field('TIME'),soft2[1].data.field('RATE'),err=soft2[1].data.field('ERROR'),gti=soft_gti2,input_counts=False)
soft_bkg2 = st.Lightcurve(sbkg2[1].data.field('TIME'),sbkg2[1].data.field('RATE'),err=sbkg2[1].data.field('ERROR'),gti=soft_gti2,input_counts=False)

hard_lc2 = st.Lightcurve(hard2[1].data.field('TIME'),hard2[1].data.field('RATE'),err=hard2[1].data.field('ERROR'),gti=hard_gti2,input_counts=False)
hard_bkg2 = st.Lightcurve(hbkg2[1].data.field('TIME'),hbkg2[1].data.field('RATE'),err=hbkg2[1].data.field('ERROR'),gti=hard_gti2,input_counts=False)

full_lc2.apply_gtis()
full_bkg2.apply_gtis()
soft_lc2.apply_gtis()
soft_bkg2.apply_gtis()
hard_lc2.apply_gtis()
hard_bkg2.apply_gtis()
```
